### understanding the dataset

Complete attribute documentation: <br>
1 id: patient identification number <br>
2 ccf: social security number (I replaced this with a dummy value of 0)  <br>
3 age: age in years  <br>
4 sex: sex (1 = male; 0 = female)  <br>
5 painloc: chest pain location (1 = substernal; 0 = otherwise)  <br>
6 painexer (1 = provoked by exertion; 0 = otherwise)  <br>
7 relrest (1 = relieved after rest; 0 = otherwise)  <br>
8 pncaden (sum of 5, 6, and 7)  <br>
9 cp: chest pain type  <br>
```
    -- Value 1: typical angina 
    -- Value 2: atypical angina 
    -- Value 3: non-anginal pain 
    -- Value 4: asymptomatic ```
10 trestbps: resting blood pressure (in mm Hg on admission to the hospital)  <br>
11 htn  <br>
`` 
Hypertensive heart disease refers to heart conditions caused by high blood pressure. The heart working under increased pressure causes some different heart disorders. Hypertensive heart disease includes heart failure, thickening of the heart muscle, coronary artery disease, and other conditions.``

12 chol: serum cholestoral in mg/dl  <br>
13 smoke: I believe this is 1 = yes; 0 = no (is or is not a smoker)  <br>
14 cigs (cigarettes per day)  <br>
15 years (number of years as a smoker)  <br>
16 fbs: (fasting blood sugar > 120 mg/dl) (1 = true; 0 = false)  <br>
17 dm (1 = history of diabetes; 0 = no such history)  <br>
18 famhist: family history of coronary artery disease (1 = yes; 0 = no)  <br>
19 restecg: resting electrocardiographic results  <br>
```
    -- Value 0: normal 
    -- Value 1: having ST-T wave abnormality (T wave inversions and/or ST elevation or depression of > 0.05 mV) 
    -- Value 2: showing probable or definite left ventricular hypertrophy by Estes' criteria ```
20 ekgmo (month of exercise ECG reading)  <br>
21 ekgday(day of exercise ECG reading)  <br>
22 ekgyr (year of exercise ECG reading)  <br>
23 dig (digitalis used furing exercise ECG: 1 = yes; 0 = no)  <br>
24 prop (Beta blocker used during exercise ECG: 1 = yes; 0 = no)  <br>
25 nitr (nitrates used during exercise ECG: 1 = yes; 0 = no)  <br>
26 pro (calcium channel blocker used during exercise ECG: 1 = yes; 0 = no)  <br>
27 diuretic (diuretic used used during exercise ECG: 1 = yes; 0 = no)  <br>
28 proto: exercise protocol  <br>
```
    1 = Bruce 
    2 = Kottus 
    3 = McHenry 
    4 = fast Balke 
    5 = Balke 
    6 = Noughton

    8 = bike 125 kpa min/min 
    9 = bike 100 kpa min/min 
    10 = bike 75 kpa min/min 
    11 = bike 50 kpa min/min 
    12 = arm ergometer ```
29 thaldur: duration of exercise test in minutes  <br>
30 thaltime: time when ST measure depression was noted  <br>
31 met: mets achieved  <br>
32 thalach: maximum heart rate achieved  <br>
33 thalrest: resting heart rate  <br>
34 tpeakbps: peak exercise blood pressure (first of 2 parts)  <br>
35 tpeakbpd: peak exercise blood pressure (second of 2 parts)  <br>
36 dummy  <br>
37 trestbpd: resting blood pressure  <br>

##### testing models

````python
# importe vários modelos (coloque mais para testar)
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVC
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.linear_model import LogisticRegression


classifiers = [
    KNeighborsClassifier(3),
    GaussianNB(),
    LogisticRegression(),
    SVC(),
    DecisionTreeClassifier(),
    RandomForestClassifier(),
    GradientBoostingClassifier()]
    

for clf in classifiers:
    clf.fit(X_train, y_train)
    name = clf.__class__.__name__
    
    print("="*30)
    print(name)
    
    print('****Results****')
    y_pred = clf.predict(X_test)
    print("Accuracy:", metrics.accuracy_score(y_test, y_pred))
    print("Precision:", metrics.precision_score(y_test, y_pred))
    print("Recall:", metrics.recall_score(y_test, y_pred))
````

#### GaussianNB and LogisticRegression have shown good accuracy. Between then, the first has the highest recall while the second has better precision. '\n'
#### As long as this dataset is about health issues, type one error is more critical and the machine has to learn more efficiently. Hence, GaussianNB model is the best one to be worked on.
